<h3><b>Get Microbiom Data

```python
import pandas as pd
import numpy as np
```

```python
original_micr_data = pd.read_excel('microbiom_data.xlsx')
```

```python
type_constrained_micr_data = pd.concat([original_micr_data.iloc[0:413,0].astype('string'),
                                        original_micr_data.iloc[0:413,1:].astype('float')], axis=1)
```

```python
numerical_micr_data = type_constrained_micr_data.copy()
sample = numerical_micr_data.pop('Sample')
visit = numerical_micr_data.pop('Visit')
```

```python
N = sample.apply(lambda row: float(row[0:3]))
N.name = 'N'
```

```python
na_micr_sum = (numerical_micr_data == 0).sum()
selected_otu = na_micr_sum[na_micr_sum<(numerical_micr_data.shape[0]*0.5)].index
otu = numerical_micr_data.loc[:, selected_otu]
```

```python
micr_data = pd.concat([N, visit, otu], axis=1)
micr_data.set_index('N', inplace=True)
```

<h3><b>Get Clinical Data

```python
original_clinical_data = pd.read_excel('clinical_data.xlsx')
clinical_param = [
                    'N', 'срок_берем1', 'диета_до_1визита', 'диета_срок', 'Глюкоза_нт_общая','через1час_тест', 'через2часа_тест',
                    'прибавка_m1','отеки1', 'АД_сист1', 'АД_диаст1', 'N_беременностей', 'N_родов', 'N_невынашивание','N_абортов',
                    'menses', 'АГ', 'НТГ', 'гсд_анамнез', 'СПКЯ', 'СД_у_родственников', 'пролактин', 'КОК', 'курение_до_беременности', 
                    'курение_во_время', 'возраст', 'рост', 'вес_до_берем','ИМТ', 'фрукты1', 'пирожные1', 'выпечка1', 'шоколад1',
                    'молочное_обезжир1', 'молочное_необезжир1', 'бобовые1', 'мясо1', 'сухофрукты1', 'рыба1', 'хлеб_цельное1', 'хлеб_любой1',
                    'соусы1', 'овощи1', 'овощи1_сырые', 'алкоголь1', 'сладкие_напитки1', 'кофе1', 'сосиски1', 'фрукты2', 'пирожные2',
                    'выпечка2', 'шоколад2', 'молочное_обезжир2', 'молочное_необезжир2', 'бобовые2', 'мясо2', 'сухофрукы2', 'рыба2',
                    'хлеб_цельнозерновой_2', 'хлеб_любой2', 'соусы2', 'овощи2', 'овощи2_сырые', 'алкоголь2', 'сладкие_напитки2', 'кофе2',
                    'сосиски2', 'ходьба1', 'подъем1', 'спорт1', 'ходьба2', 'подъем2', 'спорт2', 'глю_нт_V1', 'кетоны_V1', 'HbA1C_V1',
                    'инсулин_V1', 'лептин_V1', 'ФР_V1', 'БОМК_визит1', 'Хол_V1', 'ТГ_V1', 'ЛПВП_V1', 'ЛПОНП_V1', 'ЛПНП_V1', 'КА_V1',
                    'АБ_бер_ть', 'АБ_до_cgms1', 'АБ_срок', 'quality_cgm1'
                ]
clinical_data = original_clinical_data.loc[(original_clinical_data['quality_cgm1']==0) | (original_clinical_data['quality_cgm1']==5), clinical_param]
clinical_data.drop(columns=['quality_cgm1'], axis=1, inplace=True)
clinical_data['диета_срок'] = clinical_data['диета_срок'].astype('float')
```

```python
cat_params = clinical_data.select_dtypes('category').columns
na_cat = clinical_data[cat_params].isna().sum()>0
clinical_data[na_cat.index.tolist()] = clinical_data[na_cat.index.tolist()].apply(lambda x: x.cat.add_categories('нет информации'))
clinical_data[na_cat.index.tolist()] = clinical_data[na_cat.index.tolist()].fillna('нет информации')
```

```python
clinical_data[cat_params] = clinical_data[cat_params].apply(lambda x: x.astype('str'))
clinical_data[cat_params] = clinical_data[cat_params].apply(lambda x: x.astype('category'))
```

<h3><b>Get monitoring Data

```python
original_monitoring_data = pd.read_excel('monitoring_data.xlsx', sheet_name='Formatted')
monitoring_data = original_monitoring_data.astype('float')
```

<h3><b>Conjugate tables

```python
data = clinical_data.merge(micr_data, on='N', how='inner')
```

```python
data = monitoring_data.merge(data, on='N', how='inner')
```

```python
data.drop(columns=['CGM_mean','CGM_above','CGM_below','CGM_range','BGb60_to_mean'], axis=1, inplace=True)
```

```python
data.dropna(subset=['BG60','BG120','BGMax','AUC60','AUC120','iAUC60','iAUC120'], inplace=True)
```

```python
data = data.loc[(data['Visit']==99) | (data['Visit']==146), :]
```
